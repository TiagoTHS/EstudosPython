Desafio: Plotar as médias, medianas e moda nas notas de LC e MT (matiplotlib linha vertical)

Desafio2: Melhorar a visualização da matriz de correlação e analisar mais detalhadamente.

Desafio3: Filtrar as notas por seu estado ou sua cidade e refazer as análises, verificando se são semelhantes ao geral.

Desafio4: Pensar sobre a correlação entre matemática e Linguagens.

```
#Desafio 1

plt.figure(figsize=(12,8))
sns.histplot(dados_sem_notas_zero, x = "NU_NOTA_MT")
plt.axvline(dados_sem_notas_zero["NU_NOTA_MT"].mean(), color="red", linewidth=4, label='Media')
plt.axvline(dados_sem_notas_zero["NU_NOTA_MT"].median(), color="green", linewidth=4, label='Mediana')
plt.axvline(dados_sem_notas_zero["NU_NOTA_MT"].mode()[0], color="yellow", linewidth=4, label='Moda')

plt.annotate('{:.2f}'.format(dados_sem_notas_zero["NU_NOTA_MT"].mean()) ,xy=(dados_sem_notas_zero["NU_NOTA_MT"].mean(), 2100), xytext=(600, 2100), arrowprops=dict(facecolor='red'))
plt.annotate('{:.2f}'.format(dados_sem_notas_zero["NU_NOTA_MT"].median()) ,xy=(dados_sem_notas_zero["NU_NOTA_MT"].median(), 2350), xytext=(430, 2900), arrowprops=dict(facecolor='green'))
plt.annotate('{:.2f}'.format(dados_sem_notas_zero["NU_NOTA_MT"].median()) ,xy=(dados_sem_notas_zero["NU_NOTA_MT"].mode()[0], 2250), xytext=(300, 2250), arrowprops=dict(facecolor='yellow'))

plt.title("Nota de Matemática")
plt.legend()
```

```
plt.figure(figsize=(12,8))
sns.histplot(dados_sem_notas_zero, x = "NU_NOTA_LC")
plt.axvline(dados_sem_notas_zero["NU_NOTA_LC"].mean(), color="red", linewidth=2, label='Media')
plt.axvline(dados_sem_notas_zero["NU_NOTA_LC"].median(), color="green", linewidth=4, label='Mediana')
plt.axvline(dados_sem_notas_zero["NU_NOTA_LC"].mode()[0], color="yellow", linewidth=2, label='Moda')

plt.annotate('{:.2f}'.format(dados_sem_notas_zero["NU_NOTA_LC"].mean()) ,xy=(dados_sem_notas_zero["NU_NOTA_LC"].mean(), 2100), xytext=(600, 2100), arrowprops=dict(facecolor='red'))
plt.annotate('{:.2f}'.format(dados_sem_notas_zero["NU_NOTA_LC"].median()) ,xy=(dados_sem_notas_zero["NU_NOTA_LC"].median(), 2350), xytext=(600, 2300), arrowprops=dict(facecolor='green'))
plt.annotate('{:.2f}'.format(dados_sem_notas_zero["NU_NOTA_LC"].median()) ,xy=(dados_sem_notas_zero["NU_NOTA_LC"].mode()[0], 2250), xytext=(300, 2250), arrowprops=dict(facecolor='yellow'))

plt.title("Nota de Linguagens e Código")
plt.legend()
```
